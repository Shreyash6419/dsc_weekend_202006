```python
import pandas as pd
import numpy as np
```

```python
#Read data and describe it

df =pd.read_csv('titanic_dataset.csv')

df.head()
```

```python
df.describe(include='all')
```

```python
#Find columns with missingdata
print(df.isna().sum())
print()
print(df.isna().any(axis=0))
```

```python
#- Print shape of dataset

print(df.shape)
```

```python
# drop columns with more than 25% missing data

t=df.dropna(axis=1,thresh=(1309*0.75))

df.dropna(axis=1,thresh=(df.shape[0]*0.75), inplace=True)

print(t.isna().sum())
print()
t
```

```python
# drop columns having independent values(which do not affect the survival rate).
df.drop(columns=["Unnamed: 0", "name"], inplace = True)
df.head()
```

```python
# check data types of all columns

print(df.dtypes)
```

```python
#- convert price to numeric

a=df.fare.str.replace("$","")
a.astype('float64')
df.fare=a.astype('float64')
print(df.dtypes)
df.head()
```

```python
#- find columns still having missing/na values and also count of missing data

print(df.isna().any(axis=0))
print()
print(df.isna().sum())
```

```python
#- fill na with mean for fare and age column column.

df.fare = df.fare.fillna(df.fare.mean())
df.fare.isna().sum()
```

```python
import numpy as np
r=np.where(df.embarked.isna())
print(r)
print(type(r))
print(r[0])
print(type(r[0]))
```

```python
#- drop rows where embarked column has na values.
r=np.where(df.embarked.isna())
t=df.drop(index=r[0])
print(r)
print()
print(t)
print(np.where(t.embarked.isna()))
print()
print(t)
```

```python
#- dump the dataframe to a csv file 'titanic_filtered.csv'.

df.to_csv('titanic_filtered.csv')
```

```python
#- for surviced column replace 0 with D and 1 with A
df.survived.replace({0:'D',1:'A'})
```

```python
#- find the frequency of different values in survived column
df.survived.value_counts()
```

```python
#- group by gender and survived and see the counts in each category

grp = df.groupby(by=['gender','survived'])
print(grp)
print()
print(grp.groups)
print()
print(grp.count())
print()
print(grp.count().index)
```

```python
#- find different pclass and no of people in each class
grp = df.groupby(by=['pclass'])

print(grp.count())
```

```python
#- find top 5 people with highest values of age. Count no of male and females in the top 5

a = df.sort_values(by=['age'],ascending=False).head()
print(a)
print()
print(a.gender.value_counts())
print()
print(df.nlargest(5,columns=['age']))
```

```python
#- find max age male and female who survived
df_sur = df[df.survived==1]
print(df_sur.head())
print()
print(df_sur.groupby(by='gender').age.max())
```

```python
#- get average age by gender

avg = df.groupby(by='gender').age.mean()

print(avg)
```

```python
#- get average age by people survived vs not-survived

print(df.groupby(by='survived').age.mean())
```
